```python
import matplotlib.pyplot as plt

# Tehnologiile si intervalul LCOE (/MWh)
technologies = [
    "Solar PV", "Onshore Wind", "Offshore Wind", "Hydro", "Geothermal",
    "Biomass", "Gas CCGT", "Coal", "Nuclear", "Battery Storage"
]

lcoe_min = [20, 25, 40, 20, 40, 50, 40, 60, 60, 50]  # valori min
lcoe_max = [60, 70, 130, 80, 120, 120, 90, 120, 200, 150]  # valori max

# Culori reprezentative pentru fiecare tip de energie
colors = [
    'gold',        # Solar
    'lightgreen',  # Onshore Wind
    'green',       # Offshore Wind
    'blue',        # Hydro
    'orange',      # Geothermal
    'brown',       # Biomass
    'grey',        # Gas CCGT
    'black',       # Coal
    'purple',      # Nuclear
    'pink'         # Battery Storage
]

plt.figure(figsize=(12, 6))

for tech, min_val, max_val, color in zip(technologies, lcoe_min, lcoe_max, colors):
    plt.plot([tech, tech], [min_val, max_val], color=color, linewidth=10)

plt.ylabel("LCOE /MWh")
plt.title("Interval LCOE pentru principalele surse de energie (Europa)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# What about the time it takes to build?

```python
import matplotlib.pyplot as plt

# Technologies and their typical build times (years)
technologies = [
    "Small Hydro", "Large Hydro", "Nuclear", "Coal", 
    "Gas CCGT", "Solar PV", "Onshore Wind", "Offshore Wind"
]

build_min = [2, 5, 7, 4, 2, 0.5, 1, 3]   # minimum years
build_max = [5, 10, 12, 7, 4, 2, 2, 5]   # maximum years

colors = [
    "blue", "navy", "purple", "black", 
    "grey", "gold", "green", "teal"
]

plt.figure(figsize=(10, 6))

for tech, mn, mx, color in zip(technologies, build_min, build_max, colors):
    plt.plot([tech, tech], [mn, mx], color=color, linewidth=8)

plt.xlabel("Construction time (years)")
plt.title("Typical Construction Time for Power Plants by Technology")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
